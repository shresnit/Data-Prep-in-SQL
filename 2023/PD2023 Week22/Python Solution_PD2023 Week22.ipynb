### Preppin' Data Challenge: 2023 Week22 - Student Attendance vs Test Scores

<br>
Input the data

```python
import pandas as pd
import numpy as np

df1 = pd.read_csv('PD2023_WK22_STUDENT_TEST_SCORES.csv')
df2 = pd.read_csv('PD2023_WK22_ATTENDANCE_FIGURES.csv')
```

<br>
First we must join the test score data set and the additional class attendance data set

```python
df = pd.merge(df1, df2, on='STUDENT_NAME', how='inner')
```

<br>
Clean Data: there’s a few spelling mistakes in the data set, 
use the data tool of your choice to correct these mistakes before we proceed. These mistakes are in the “Subject”  Column

```python
df['SUBJECT'] = np.where(df['SUBJECT'].str.contains('M'),'Math', df['SUBJECT'])
df['SUBJECT'] = np.where(df['SUBJECT'].str.contains('S'),'Science', df['SUBJECT'])
df['SUBJECT'] = np.where(df['SUBJECT'].str.contains('E'),'English', df['SUBJECT'])
```

<br>
Use a calculated field to clean up the test score column, so each number is a rounded to the nearest whole number

```python
#Rounding to nearest whole number
df['TestScoreInteger'] = df['TEST_SCORE'].round()

#Changing to integer
df['TestScoreInteger'] = df['TestScoreInteger'].astype(int)
```

<br>
Split the student name column so we can see the first name and surname in two separate columns, named appropriately

```python
df[['First Name', 'SurName']] = df['STUDENT_NAME'].str.split('_', expand=True)
```

<br>
If any student has an attendance percentage less than 70%, flag it as "Low Attendance" in a new column "attendance flag". Conversely, if above 90%, flag as “high attendance”. And anything else as “Medium”.

```python
df['Attendance Flag'] = np.where(df['ATTENDANCE_PERCENTAGE'] < 0.7, 'Low Attendance', ' ')
df['Attendance Flag'] = np.where(df['ATTENDANCE_PERCENTAGE'] > 0.9, 'High Attendance', df['Attendance Flag'] )
df['Attendance Flag'] = np.where(df['Attendance Flag'] == ' ', 'Medium', df['Attendance Flag'])
```

<br>
Output the data

```python
#Arranging column order as per the desired output
df = df[[
        'Attendance Flag', 
        'First Name', 
        'SurName', 
        'ATTENDANCE_PERCENTAGE', 
        'STUDENT_ID', 
        'SUBJECT', 
        'TEST_SCORE',
        'TestScoreInteger' 
        ]]

#Output
df.to_csv('PD2023_W22_Output.csv', index=False)
```
