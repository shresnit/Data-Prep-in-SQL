### Preppin' Data Challenge 2024: Week 3 Performance Against Targets
Input the outputs from 2024 Week 1 challenge

```python
import pandas as pd
import numpy as np

df1 = pd.read_csv('PD2024_WK1_OUTPUT_FLOW_CARD.csv')
df2 = pd.read_csv('PD2024_WK1_OUTPUT_NONFLOWCARD.csv')

df = pd.concat([df1, df2], ignore_index=True)
```

<br>
Input the new targets Excel sheet (Q1 - 4) 

```python
df3 = pd.read_excel('PD2024_WK3_INPUT_TARGETS.xlsx', sheet_name=None)
df3 = pd.concat([df3['Q1'], df3['Q2'], df3['Q3'],df3['Q4']], ignore_index=True)

df3['FirstLetter'] = df3['Class'].str[0]
df3 = df3.rename(columns = {'Month': 'Date'})
```

<br>
Correct the Classes being incorrect as per last week

- Economy to First
- First Class to Economy
- Business Class to Premium
- Premium Economy to Business

```python
df['Class'] = df['Class'].apply(lambda x: 'First' if x == 'Economy'
                               else 'Economy' if x =='First Class'
                               else 'Premium' if x =='Business Class'
                               else 'Business')
```

<br>
Find the First Letter from each word in the Class to help with joining the Targets data to Sales data

```python
df['FirstLetter'] = df['Class'].str[0]
```

<br>
Change the date to a month number

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df['Date'] = df['Date'].dt.month
```

<br>
Total up the sales at the level of:

- Class
- Month

```python
df = df.groupby(['Date', 'FirstLetter'], as_index=False).agg({'Price': sum})
```

<br>
Join the Targets data on to the Sales data (note - you should have 48 rows of data after the join)

```python
df = pd.merge(df,df3, on=['FirstLetter', 'Date'], how='inner')
```

<br>
Calculate the difference between the Sales and Target values per Class and Month

```python
df['Difference to Target'] = df['Price'] - df['Target']
```

<br>
Output the data

```python
#reordering columns as desired output
df = df[['Difference to Target', 'Date', 'Price', 'Class', 'Target']]
```

```python
df.to_csv('PD2024W3_OUTPUT.csv', index=False)
```
