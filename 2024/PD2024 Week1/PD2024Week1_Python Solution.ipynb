Input the data

```python
import pandas as pd
import numpy as np

file = 'PD2024W1.csv'
df = pd.read_csv(file)
```

Split the Flight Details field to form:
- Date 
- Flight Number
- From
- To
- Class
- Price

```python
# Splitting Flight_details in Data and Flight Number
df['Date'] = df['Flight_Details'].str.split('//', expand=True)[0]
df['Flight Number'] = df['Flight_Details'].str.split('//', expand=True)[1]

# Splitting Flight_details into 'From-To' with '//' delimeter and then splitting again the 'From-To'
# into 'From' and 'To' with '-' delimeter
df['From'] = (df['Flight_Details'].str.split('//', expand=True)[2]).str.split('-', expand=True)[0]
df['To'] = (df['Flight_Details'].str.split('//', expand=True)[2]).str.split('-', expand=True)[1]

# Splitting Flight_details in Class and Price
df['Class'] = df['Flight_Details'].str.split('//', expand=True)[3]
df['Price'] = df['Flight_Details'].str.split('//', expand=True)[4]
```

<br>
Convert the following data fields to the correct data types

- Date to a date format
- Price to a decimal value

```python
# Converting Date to Date format

df['Date'] = pd.to_datetime(df['Date'], format = '%Y-%m-%d')
```

```python
# Converting Price to a decimal value
df['Price'] = df['Price'].astype(float)
df['Price'] = ["{:.2f}".format(x) for x in df['Price']]
```

<br>
Change the Flow Card field to Yes / No values instead of 1 / 0

```python
df['Flow Card?'] = np.where(df['Flow_Card'] == 1 , 'Yes', 'No')
```

<br>
Create two tables, one for Flow Card holders and one for non-Flow Card holders

```python
# Ordering the table as per the desired output
column_order = ['Date', 'Flight Number', 'From', 'To', 'Class', 'Price', 'Flow Card?', 'Bags_Checked', 'Meal_Type']
df = df[column_order]
```

```python
df_Flow_Card_Holders = df[df['Flow Card?'] == 'Yes']
df_non_Flow_Card_Holders = df[df['Flow Card?'] == 'No']
```

<br>
Output the data sets

```python
df_Flow_Card_Holders.to_csv('PD2024W1_Output1.csv', index = False)
df_non_Flow_Card_Holders.to_csv('PD2024W1_Output2.csv', index = False)
```
